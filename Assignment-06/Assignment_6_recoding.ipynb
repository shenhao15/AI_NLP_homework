```python
x_train = tf.cast(x_train[...,tf.newaxis]/255, tf.float32),
x_test = tf.cast(x_test[...,tf.newaxis]/255, tf.float32),

#y_train = y_train.astype('float32')
#y_test = y_test.astype('float32')
y_train = tf.keras.utils.to_categorical(y_train,10)
y_test = tf.keras.utils.to_categorical(y_test,10)
```

```python
y_test[:2]
```

```python
y_train[3]
```

```python
#Build the model using Sequential

mnist_model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32,[3,3],activation='relu',
                          input_shape=(28,28,1)),
    tf.keras.layers.Conv2D(64,[3,3],activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128,activation="relu"),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(10,activation="softmax")
])
```

```python
mnist_model.summary()
```

```python
mnist_model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=1e-3),
                    validation_split=0.1,shuffle=True,
                   loss = tf.keras.losses.categorical_crossentropy,
                   metrics = ["accuracy"])
```

```python
mnist_model.fit(x_train,y_train,batch_size=128,epochs=10)
```

```python
mnist_model.fit(x_train,y_train,batch_size=128,epochs=100)
```

```python
mnist_model.evaluate(x_test,y_test)
```

```python
#Build the model using Model
inputs = tf.keras.Input(shape=(28,28,1),name="digits")
conv_1 = tf.keras.layers.Conv2D(32,[3,3],activation="relu")(inputs)
conv_2 = tf.keras.layers.Conv2D(64,[3,3],activation="relu")(conv_1)
max_pool = tf.keras.layers.MaxPooling2D(pool_size=(2,2))(conv_2)
dropout = tf.keras.layers.Dropout(0.25)(max_pool)
flatten = tf.keras.layers.Flatten()(dropout)
dense_1 = tf.keras.layers.Dense(128,activation="relu")(flatten)
dropout_2 = tf.keras.layers.Dropout(0.5)(dense_1)
outputs = tf.keras.layers.Dense(10,activation="softmax")(dropout_2)

mnist_model_3 = tf.keras.Model(inputs=inputs,outputs=outputs)
```

```python
mnist_model_3.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=1e-3),
                    validation_split=0.1,shuffle=True,
                   loss = tf.keras.losses.categorical_crossentropy,
                   metrics = ["accuracy"])
```
